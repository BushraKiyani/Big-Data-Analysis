```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. Loading the Training Data

```python
df = pd.read_csv("synthetic_dataset.csv")
df
```

# 2. Data Cleaning and Preprocessing
## 2.1  Dealing with Missing Values
We'll check the null values in our dataset.

```python
# Check NaN values
df.info()
df.isna().sum()
```

```python
df.isnull().any(axis=1).sum()
```

### 2.1.1 Droping rows containing NaN
We see among 4500 observations we have 332 observations with null values. So first we'll implement classification algorithms by droping these 332 observations/rows.

```python
df_DropNa = df.dropna()
df_DropNa.info()
```

```python
for col in df_DropNa:
    print(df_DropNa[col].nunique())
```

```python
for col in df_DropNa:
    print("name of Column is:", col , sorted(df_DropNa[col].unique()))
```

```python
data_Encoded = pd.get_dummies(df_DropNa, columns=["Feature_J", "Label"], drop_first=True)
data_Encoded = pd.get_dummies(data_Encoded)
data_Encoded
```

# 3. Spliting into Training + Validation

```python
import sklearn.model_selection
train,validation = sklearn.model_selection.train_test_split(data_Encoded, test_size=0.1,random_state=0)
train.head(5)
```

# 4. Learning the decision tree

```python
import sklearn.tree
dt = sklearn.tree.DecisionTreeClassifier(min_samples_leaf=50)
X, y = train.drop("Label_Yes",axis=1), train["Label_Yes"]
dt.fit(X, y)
```

```python
from sklearn.tree import export_graphviz
export_graphviz(dt, "tree.dot", feature_names=X.columns, class_names=["n","y"], filled=True)
```

```python
#from subprocess import check_call
#check_call(['dot','-Tpng',"tree.dot",'-o','tree.png'])
```

```python
import sklearn.metrics
print("Decision Tree Training accuracy: ", end="")
print(sklearn.metrics.accuracy_score(train["Label_Yes"], dt.predict(train. drop("Label_Yes",axis=1))))
print("Decision Tree Validation accuracy:", end="")
print(sklearn.metrics.accuracy_score(validation["Label_Yes"], dt.predict(validation.drop("Label_Yes",axis=1))))
```
