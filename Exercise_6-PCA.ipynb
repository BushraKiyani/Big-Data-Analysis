```python
import numpy as np
import pandas as pd

from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('fifa_20.csv')#Change path according to your files
```

### PCA

A common method for dimensionality reduction is the Principal Component Analysis, which is also known from the lecture:

https://en.wikipedia.org/wiki/Principal_component_analysis

Create scatter plots of the features 'Dribbling' and 'Ball control', then apply z-normalization to both features, and finally applying PCA to the normalized 2-d data (keep both dimensions after the transformation).

What is the percentage of variance explained by the remining component, if we reduce the dimensionality of the 2-d dataset to 1-d with PCA?

## 1. Skill Dribbling Vs Skill Ball Control

```python
#TODO: make scatter plots of the original normalized and the transformed data
plt.title("Scatter Plot: Skill Dribbling Vs Skill Ball Control", size = 14, pad = 10)
plt.xlabel("Skill Dribbling")
plt.ylabel("Skill Ball Control")
plt.scatter(df['skill_dribbling'],df['skill_ball_control'], c= "red")
plt.show()

scale_dribbling = scale(df['skill_dribbling'], axis=0, with_mean=True, with_std=True, copy=True)
scale_ballControl = scale(df['skill_ball_control'], axis=0, with_mean=True, with_std=True, copy=True)
plt.title("Scatter Plot: Scaled Skill Dribbling Vs Scaled Skill Ball Control", size = 14, pad = 10)
plt.xlabel("Scaled Skill Dribbling")
plt.ylabel("Scaled Skill Ball Control")
plt.scatter(scale_dribbling ,scale_ballControl, c= "green")
plt.show()
```

```python
# TODO: output the amount of explained variance of each component and remining variance after dimensionality reduction

scaledData= np.column_stack((scale_dribbling, scale_ballControl))
df2 = pd.DataFrame(scaledData).dropna()
pca = PCA(n_components= 2, copy=True, whiten=False, svd_solver='auto', tol=0.0, iterated_power='auto', random_state=None)
pca.fit(df2)
print("Variance explained by Skill Dribbling and Skill Ball Controll :", pca.explained_variance_)


# remining variance after dimensionality reduction


pca = PCA(n_components= 1, copy=True, whiten=False, svd_solver='auto', tol=0.0, iterated_power='auto', random_state=None)
pca.fit(df2)

print("Percentage of Variance explained after dimensionality reduction:", pca.explained_variance_ratio_)
```

## 2. Dribbling Vs SKIl Ball Control


```python
#TODO: make scatter plots of the original normalized and the transformed data
plt.title("Scatter Plot: Dribbling Vs Ball Control", size = 14, pad = 10)
plt.xlabel("Dribbling")
plt.ylabel("Ball Control")
plt.scatter(df['dribbling'],df['skill_ball_control'], c= "red")
plt.show()

scale_dribbling = scale(df['dribbling'], axis=0, with_mean=True, with_std=True, copy=True)
scale_ballControl = scale(df['skill_ball_control'], axis=0, with_mean=True, with_std=True, copy=True)
plt.title("Scatter Plot: Scaled Dribbling Vs Scaled Ball Control", size = 14, pad = 10)
plt.xlabel("Scaled Dribbling")
plt.ylabel("Scaled Ball Control")
plt.scatter(scale_dribbling ,scale_ballControl, c= "green")
plt.show()
```

```python
# TODO: output the amount of explained variance of each component and remining variance after dimensionality reduction

scaledData= np.column_stack((scale_dribbling, scale_ballControl))
df2 = pd.DataFrame(scaledData).dropna()
pca = PCA(n_components= 2, copy=True, whiten=False, svd_solver='auto', tol=0.0, iterated_power='auto', random_state=None)
pca.fit(df2)
print("Variance explained by Dribbling and Skill Ball Controll :", pca.explained_variance_)


# remining variance after dimensionality reduction


pca = PCA(n_components= 1, copy=True, whiten=False, svd_solver='auto', tol=0.0, iterated_power='auto', random_state=None)
pca.fit(df2)
print("Percentage of Variance explained after dimensionality reduction:", pca.explained_variance_ratio_)
```
